### Exercise 1: Line Plot Customization

```python
# TODO: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.linspace(0, 10, 100)

# Calculate y values as the square of x
y = x**2

# Create the plot
plt.plot(x, y)

# Customize the plot
plt.title('Square of x (y = x^2)')
plt.xlabel('x values')
plt.ylabel('y values (x^2)')
plt.grid(True)

# Show the plot
plt.show()
```
